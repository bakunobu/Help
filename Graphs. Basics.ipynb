# Simple graphs

Create two lists of ints

```python
x_num = [1, 2, 3]
y_num = [2, 4, 6]
# And the points are: (1, 2), (2, 4), (3, 6)
```

```python
from pylab import plot, show
%matplotlib tk #the other option is inline (don't have qt installed).
# if you have ImportError: No module named 'tkinter' on Ubuntu, use sudo apt-get install python3-tk first
```

```python
plot(x_num, y_num) # creates graph first arg - X axis, second - Y axis
show() #displays graph
```

marker keyword argument - changes points style
Available:
'.'
'o'
'^'
'+'
'x'

```python
%matplotlib inline
```

Use 'marker' keyword argument to make dots visible on graph. By adding just a dot sign you display only dots (no graph line)

```python
plot(x_num, y_num, marker='^')#marker keyword arg
show()
```

```python
plot(x_num, y_num, '^')# sign as an arg
show()
```

if there's only one data argument it's y-axis by default (x-axis is filled with values' indexes)

```python
nyc_temp = [53.9, 56.3, 56.4, 53.4, 54.5, 55.8, 56.8, 55.0, 55.3, 54.0, 56.7, 56.4, 57.3]
plot(nyc_temp, marker='o')
```

```python
years = range(2000,2013)
plot(years, nyc_temp, marker='o') # second data arg is added
show()
```

    Graphs comparison

```python
nyc_temp_2000 = [31.3, 37.3, 47.2, 51.0, 63.5, 71.3, 72.3, 72.7, 66.0, 57.0, 45.3, 31.1]
nyc_temp_2006 = [40.9, 35.7, 43.1, 55.7, 63.1, 71.0, 77.9, 75.8, 66.6, 56.2, 51.9, 43.6]
nyc_temp_2012 = [37.3, 40.9, 50.9, 54.8, 65.1, 71.0, 78.8, 76.7, 68.8, 58.0, 43.9, 41.5]
```

```python
months = range(1, 13)
```

```python
plot(months, nyc_temp_2000, months, nyc_temp_2006, months, nyc_temp_2012)
```

We can do the same by creating saparate plots

```python
from pylab import legend
```

```python
plot(months, nyc_temp_2000)
plot(months, nyc_temp_2006)
plot(months, nyc_temp_2012)
legend([2000, 2006, 2012],)#names can be typed as plot(args)
show()
```

Label position can be adjusted by the lock= argument
Available options:
'best',
'lower center',
'center left',
'upper left' (default)

Graph customization
title() - for the title
xlabel() - adding the x-axis label
ylabel() - adding the y-axis label

```python
from pylab import plot, show, title, xlabel, ylabel
```

```python
plot(months, nyc_temp_2000)
plot(months, nyc_temp_2006)
plot(months, nyc_temp_2012)
legend([2000, 2006, 2012],)
title('Average monthly temperature in NYC')
xlabel('Month')
ylabel('Temperature')
show()
```

An axis range adjustment

```python
from pylab import axis
```

```python
plot(years, nyc_temp, marker='o')
title('Average annual temperature in NYC')
xlabel('Year')
ylabel('Temperature')
axis(ymin=25, ymax=75) # set borders (xmin and xmax for x-axis as well)
show()
```

## Saving plots

We use savefig() functions to save plots (PNG, PDF and SVG are available. Specify while adding an extention to the file's name)

```python
plt.plot(years, nyc_temp, marker='o')
plt.title('Average annual temperature in NYC')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.axis(ymin=25, ymax=75)
plt.savefig('NYC_weather.png')
```
