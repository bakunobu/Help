# Newton's Law of Universal Gravitation

\begin{align}
F = \frac{Gm_1m_2}{r^2}
\end{align}

```python
import scipy.constants as const #Need the Newtonian constant of gravitation from this library
#(G or gravitational_constant)
```

```python
import matplotlib.pyplot as plt
```

```python
def draw_graph(x, y):
    plt.plot(x, y, marker='o')
    plt.xlabel('Distance in meters')
    plt.ylabel('Gravitational force in newtons')
    plt.title('Gravitational force and distance')
    plt.show()
    
    
def newt_force_calc(a, b, dist):
    """a and b are masses (int or float)
    dist is a tuple or a list;
    the function returns a list of forces for each distance"""
    return list((const.G * m_1 * m_2) / (r ** 2) for r in dist)

    
m_1, m_2 = 0.5, 1.5 #masses
dist = list(x for x in range(100,1001, 50)) #list of distances


force = newt_force_calc(m_1, m_2, dist)
draw_graph(dist, force)
```

# Projectile motion

Модуль вектора скорости по оси **X**
\begin{align}
u_x = u\cos\theta
\end{align}

Модуль вектора скорости по оси **Y**
\begin{align}
u_y = u\sin\theta
\end{align}

\begin{align}
S_x = u\cos\theta t
\end{align}

\begin{align}
S_y = u\cos\theta - \frac{1}{2}gt^2
\end{align}

```python
import math
```

```python
def draw_tr_graph(x, y):
    plt.plot(x, y)
    plt.xlabel('x-coordinate')
    plt.ylabel('y-coordinate')
    plt.title('projectile motion of a ball')
```

```python
def timing(start=0.0, stop=0.2, increment = 0.001):
    intervals = []
    while start < stop:
        start += increment
        intervals.append(start)
    return intervals
```

```python
def draw_trajectory(u, theta):
    theta = math.radians(theta)
    t_flight = 2 * u * math.sin(theta) / const.g # const.g is standart accceleration of gravity
    intervals = timing(0, t_flight, 0.001)
    x = list(u * math.cos(theta) * t for t in intervals)
    y = list(u * math.sin(theta) * t - 0.5 * const.g * (t ** 2) for t in intervals)
    draw_tr_graph(x, y)
```

```python
def main():
    if __name__ == '__main__':
        try:
            u = float(input('Enter the initial velocity (m/s): '))
            theta = float(input('Enter the angle of projection (degrees): '))
        except ValueError:
            print('Your input is incorrect!')
        else:
            draw_trajectory(u, theta)
            plt.show()
```

```python
main()
```

```python
def multi_velocity(my_list):
    if __name__ == '__main__':
        try:
            theta = float(input('Enter the angle of projection (degrees): '))
        except ValueError:
            print('Your input is incorrect!')
        else:
            for u in my_list:
                draw_trajectory(u, theta)
        leg_list = list(map(lambda x: 'vel = ' + str(x), my_list))
        plt.legend(leg_list)
        plt.show()
```
