## Task 3. Enhanced Projectile Trajectory Comparison Program

```python
import scipy.constants as const
import math
```

```python
def draw_tr_graph(x, y):
    plt.plot(x, y)
    plt.xlabel('x-coordinate')
    plt.ylabel('y-coordinate')
    plt.title('projectile motion of a ball')
```

```python
def draw_trajectory(u, theta):
    theta = math.radians(theta)
    t_flight = 2 * u * math.sin(theta) / const.g # const.g is standart accceleration of gravity
    intervals = timing(0, t_flight, 0.001)
    x = list(u * math.cos(theta) * t for t in intervals)
    y = list(u * math.sin(theta) * t - 0.5 * const.g * (t ** 2) for t in intervals)
    return x, y
```

```python
def timing(start=0.0, stop=0.2, increment = 0.001):
    intervals = []
    while start < stop:
        start += increment
        intervals.append(start)
    return intervals
```

```python
def how_many_calc(text):
    calc = 'Ni!'
    while calc == 'Ni!':
        try:
            calc = int(input(text))
            return calc
        except ValueError:
            print('Must be an integer!')
```

```python
def main():
    if __name__ == '__main__':
        calc = how_many_calc('How many calculations do you need?: ')
        param = []
        while len(param) < calc:
            try:
                u = float(input('Enter the initial velocity (m/s): '))
                theta = float(input('Enter the angle of projection (degrees): '))
                param.append((u, theta))
            except ValueError:
                print('Your input is incorrect!')
            else:
                
                all_res = []
                leg = []
                for u, theta in param:
                    leg.append('vel = ' + str(u) + ' & ' + str(theta))
                    x, y = draw_trajectory(u, theta)
                    all_res.append((max(x), max(y)))
                    plt.plot(x, y)
        
        plt.xlabel('x-coordinate')
        plt.ylabel('y-coordinate')
        plt.title('projectile motion of a ball')
        plt.legend(leg)
        for i in range(0,len(param)):
            print('For {0} \nmax x: {1:.3f} \nmax y: {2:.3f}\n'.format(leg[i], all_res[i][0], all_res[i][1]))
        plt.show()
```

```python
main()
```

## Task 4. Expenses Visualization

```python
def create_bar_chart(expenses, categories):
    num_bars = len(expenses)
    positions = range(1, num_bars +1)
    plt.barh(positions, expenses, align='center')
    plt.yticks(positions, categories)
    
```

```python
def bar_main():
    if __name__ == '__main__':
        calc = how_many_calc('How many categories do you need?: ')
        categories = []
        expenses = []
        while len(categories) < calc:
            try:
                expen = input('Enter the category ')
                amount = float(input('How much did you spend?: '))
                categories.append(expen)
                expenses.append(amount)
                
            except ValueError:
                print('Your input is incorrect!')
            else:
                create_bar_chart(expenses, categories)
    plt.xlabel('Expenses')
    plt.ylabel('Categories')
    plt.title('Weekly Expenses')
    plt.grid()
    plt.show()
```

```python
bar_main()
```

## The Fibonacci Sequence and the Golden Ratio

```python
def create_fib_seq(stop):
    a, b = 1, 1
    fib_seq = []
    while len(fib_seq) < stop:
        a, b = b, a + b
        fib_seq.append(b)
    return fib_seq
```

```python
def n_ratio(my_list):
    ratios = [my_list[i] / my_list[i-1] for i in range(1, len(my_list))]
    return ratios
```

```python
def fib_ratio_plot(my_list):
    num_of_el = [x for x in range(1, len(my_list)+1)]
    plt.plot(num_of_el, my_list)
    plt.xlabel('Number_of_elements')
    plt.ylabel('Ratio')
    plt.title('The Fibonacci Sequece and the Golden Ratio')
    plt.show()
```

```python
fib = create_fib_seq(100)
rat = n_ratio(fib)
fib_ratio_plot(rat)
```
